```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = pd.merge(city,ride)

# Display the data table for preview
df = pd.DataFrame(df)


df.head()
```

```python
average_fare = df.groupby(['city']).mean()['fare']
total_rides = df.groupby(['city']).count()['fare']
total_drivers = city.groupby(['city']).sum()['driver_count']

city_type = city.set_index("city")["type"]

pyber_df = pd.DataFrame({"Average Fare" : average_fare, "Total Rides" : total_rides,
                         "Total Drivers" : total_drivers, "City Type" : city_type})
pyber_df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
rural = pyber_df[pyber_df["City Type"] == "Rural"]
suburban = pyber_df[pyber_df["City Type"] == "Suburban"]
urban = pyber_df[pyber_df["City Type"] == "Urban"]

```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types

plt.figure(figsize = (10,7))

r = plt.scatter(rural["Total Rides"], rural["Average Fare"], s = rural["Total Drivers"]*15, alpha = .65,
           color = "gold", edgecolors = "red", linewidths = 1.5, label = "Rural")

s = plt.scatter(suburban["Total Rides"], suburban["Average Fare"], s = suburban["Total Drivers"]*15, alpha = .65,
           color = "lightskyblue", edgecolors = "green", linewidths = 1.5, label = "Suburban")

u = plt.scatter(urban["Total Rides"], urban["Average Fare"], s = urban["Total Drivers"]*15, alpha = .65,
           color = "lightcoral", edgecolors = "black", linewidths = 1.5, label = "Urban")


# Incorporate the other graph properties
plt.suptitle("Pyber Ride Sharing Data")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Ride Fare ($)")

plt.grid(True)
# Create a legend

plt.legend(handles=[r,s,u], prop= {"size":15})
# Incorporate a text label regarding circle size
plt.title("*Bubble size corrrelates to Number of Drivers per City*")
# Save Figure
```

```python
# Show plot
plt.show()
```
